```python
import argparse
import sys

from tensorflow.examples.tutorials.mnist import input_data

import tensorflow as tf
```

```python
# import data
mnist = input_data.read_data_sets('.')
```

```python
# create the model

# input
x = tf.placeholder(tf.float32, [None, 784])
```

```python
# weights and biases
W1 = tf.Variable(tf.random_uniform([784, 10]))
b1 = tf.Variable(tf.zeros([10]))
W2 = tf.Variable(tf.random_uniform([10, 10]))
b2 = tf.Variable(tf.zeros([10]))
```

```python
# output of 1st hidden layer
# linear transform 784 -> 10 :: ReLU
h1 = tf.matmul(

        x,
        W1

    ) + b1

h1_relu = tf.nn.relu(h1)

# linear transform 10 -> 10
y = tf.matmul(h1_relu, W2) + b2

# Placeholder for the labels
y_ = tf.placeholder(tf.int64, [None])
```

```python
# define the loss and set up SGD 
cross_entropy = tf.losses.sparse_softmax_cross_entropy(labels=y_, logits=y)
train_step = tf.train.GradientDescentOptimizer(0.5).minimize(cross_entropy)
```

```python
# sess is a tensorflow session.
sess = tf.InteractiveSession()
```

```python
# initialize variables
tf.global_variables_initializer().run()
```

```python
# Train for 10 epochs, 1000 batches per epoch
for epoch in range(10):
    for batch_id in range(10000):
        batch_xs, batch_ys = mnist.train.next_batch(100)
        # run the optimizer
        sess.run(train_step, feed_dict={x: batch_xs, y_: batch_ys})
        # compute the loss each batch
        loss = sess.run(cross_entropy, feed_dict={x: batch_xs, y_: batch_ys})

        if batch_id % 10 == 0:
            print('Batch:', batch_id, 'Loss:', loss)
```
