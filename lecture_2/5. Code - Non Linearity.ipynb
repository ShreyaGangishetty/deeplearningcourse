```python
import numpy as np
circle_xs = np.load('circles_xs.npy')
circle_one_hot = np.load('circle_one_hot.npy')
print(circle_xs.shape, circle_one_hot.shape)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.scatter(circle_xs[:,0], circle_xs[:,1])
```

```python
import keras # import the framework
from keras.models import Sequential
from keras.layers import Dense, Activation

model = Sequential()
model.add(Dense(10, input_dim=2))
model.add(Activation('relu'))
model.add(Dense(2))

# linear separator bits
model.add(Activation('softmax'))
model.compile(loss='categorical_crossentropy', # a standard measure of error for 2-category problems, used for knob tweaking
              optimizer='sgd', # this is the knob-tuning portion. also standard.
              metrics=['accuracy']) # this is just so we can see how good the predictions are

model.fit(circle_xs, circle_one_hot, epochs=3000)
```
