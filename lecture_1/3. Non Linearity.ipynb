```python
# general imports
import numpy as np
import matplotlib.pyplot as plt
```

```python
# load data
circle_xs = np.load('circles_xs.npy')
circle_ys = np.load('circles_ys.npy')

outer_circle = circle_xs[np.where(circle_ys==1)]
inner_circle = circle_xs[np.where(circle_ys==0)]
```

```python
# plot both circles
plt.scatter(inner_circle[:,0], inner_circle[:,1], c='r')
plt.scatter(outer_circle[:,0], outer_circle[:,1], c='b')
```

```python
# plot both circles
inner_zs = inner_circle[:,0] ** 2 + inner_circle[:,1] ** 2
outer_zs = outer_circle[:,0] ** 2 + outer_circle[:,1] ** 2

from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(inner_circle[:,0], inner_circle[:,1], inner_zs, c='r')
ax.scatter(outer_circle[:,0], outer_circle[:,1], outer_zs, c='b')
```

```python
# plot both circles
inner_zs = inner_circle[:,0] ** 2 + inner_circle[:,1] ** 2
outer_zs = outer_circle[:,0] ** 2 + outer_circle[:,1] ** 2

from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(inner_circle[:,0], inner_circle[:,1], inner_zs, c='r')
ax.scatter(outer_circle[:,0], outer_circle[:,1], outer_zs, c='b')

# all this bullshit is to draw a plane
ax = plt.gca()
X = np.arange(-1,1,0.1)
Y = np.arange(-1,1,0.1)
X, Y = np.meshgrid(X, Y)
Z = np.ones((20, 20)) * 0.82

ax.plot_surface(X, Y, Z, alpha=0.2, cmap=cm.coolwarm, linewidth=0, antialiased=False)
```
