```python
# first, load the data
import numpy as np
r_set = np.load('r_set.npy')
b_set = np.load('b_set.npy')
```

```python
# xs:
inputs = np.concatenate([r_set, b_set])
one_hot_labels = np.load('r_b_one_hot.npy')
print(inputs.shape)
print(one_hot_labels)
```

```python
import keras # import the framework
from keras.models import Sequential # sequential makes it easy to specify the architecture
from keras.layers import Dense, Activation # Dense is keras' name for the non-linear building block

model = Sequential()
model.add(Dense(2, input_dim=2))

# linear separator bits
model.add(Activation('softmax'))
model.compile(loss='categorical_crossentropy', # a standard measure of error for 2-category problems
              optimizer='sgd',
              metrics=['accuracy']) # this is just so we can see how good the predictions are

model.fit(inputs, one_hot_labels, epochs=6)
```
